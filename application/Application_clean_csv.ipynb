```python
def transform_float_to_int(df, columns):
    int_column = []
    for column in columns:
        column = [int(df[column][i]) for i in range(len(df[column]))]
        int_column.append(column)
    print(int_column)
    return int_column

columns = ['sl_nb_chambre', 'sl_taille','sl_salle_d_eau','sl_salle_de_bain',
           'sl_toilette', 'sl_etage', 'sl_hauteur', 'sl_parking', 'sl_terrasse', 'sl_nb_piece']

new_columns_1 = transform_float_to_int(paris, columns)
new_columns_2 = transform_float_to_int(lyon, columns)
```

```python
def create_dataframe(new_df, dataset, df, columns):
    new_df = pd.DataFrame(dataset)
    new_df = new_df.transpose()
    new_df = new_df.rename(columns={0: "sl_nb_chambre", 
                      1: "sl_taille",
                      2: "sl_salle_d_eau",
                      3: "sl_salle_de_bain",
                      4: "sl_toilette",
                      5: "sl_etage",
                      6: "sl_hauteur",
                      7: "sl_parking",
                      8: "sl_terrasse",
                      9: "sl_nb_piece"})
    df = df.drop(columns, axis=1)
    res = pd.concat([df, new_df], axis=1, sort=False)
    return res
    
paris_df = create_dataframe(paris_1, new_columns_1, paris, columns)
lyon_df = create_dataframe(lyon_1, new_columns_2, lyon, columns)
```
