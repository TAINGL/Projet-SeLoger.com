# Relation between target and features

```python
def filter_corr(data, target):
    df_corr = data.corr()
    df = (df_corr[df_corr[target] >= 0.4])
    return df[target]

filter_corr(data, 'sl_prix')
```

```python
result = pd.DataFrame(filter_corr(data, 'sl_prix')) 
corr_list = list(result.index)
corr_list = list(result.index)
ax = sns.heatmap(data[corr_list].corr(),annot=True)
```

# Null Values 

```python
def viz_na(data):
    return sns.heatmap(data.isna(), cbar=False)

viz_na(data)
# on a pas beaucoup de valeur manquante, les features ont des relations en communs 
```

```python
#missing data
def nan_description(df):
    total = df.isnull().sum().sort_values(ascending=False)
    percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    return missing_data.head(20)

nan_description(data)  
```
